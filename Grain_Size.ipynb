#Note
###This python code takes only one input (grain size in millimeters)
### The output is:


1.   The input grain size value in mm (Wentworth scale value)
2.   Krumbein's scale value
3.   Udden Grade scale value
4.   Udden-Wentworth Nomenclature
5.   Some Details

```python
import math as m

n = float(input("Enter the grain size in mm: "))

print(f"Input grain size (True Nominal diameter): {n} mm")

kphi = m.log2(n)*-1 #gives Krumbeins phi-scale value
print("Krumbein's scale value: ", kphi)
print(f"Udden grade scale value: 2^{-kphi}")
print("Udden-Wentworth Nomenclature: ")

if  kphi <= -8.0:
    print("Boulder")
elif -8.0< kphi <= -6.0:
    print("Cobble")
elif -6.0 < kphi <= -2.0:
    print("Pebble")
elif -2.0 < kphi <= -1.0:
    print("Granule")
elif -1.0 < kphi <= 0.0:
    print("Sand, Very coarse sand")
elif 0.0 < kphi <= 1.0:
    print("Sand, Coarse sand")
elif 1.0 < kphi <= 2.0:
    print("Sand, Medium sand")
elif 2.0 < kphi <= 3.0:
    print("Sand, Fine sand")
elif 3.0 < kphi <= 4.0:
    print("Sand, Very Fine sand")
elif 4.0 < kphi <= 8.0:
    print("Silt")
elif kphi > 8.0:
    print("Clay")
if kphi <= -8.0 or kphi < -1.0:
  print("Details: Rudaceous Rocks\n Sediment: Gravel\n Rudites rocks. eg: Conglomerates, Breccias")  
if -1.0 <= kphi <= 4.0:
  print("Details: Arenaceous rocks\n Sediment: Sand\n Sandstones. eg: Arenites and Wackes")
if kphi > 4.0:
  print("Details: Argillaceous Rocks\n Sediment: Mud\n Lutites (Mudrocks)")   
```
